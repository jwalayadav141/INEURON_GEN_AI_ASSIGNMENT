
I have created two new column for remove stopwords as "text_without_stopword" and for stemming as "stem" and for token as "tokenize"

```python
import os
import pandas as pd
import nltk
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/jwalayadav141/INEURON_GEN_AI_ASSIGNMENT/main/sample.csv")
```

Conver to lower case

```python
data['text']=data['text'].str.lower()
```

```python
for word in data['text']:
  print(word)
```

Remove the url

```python
import re
def remove_url(text):
    pattern=re.compile("https?://[a-zA-Z]*.[a-zA-Z]*.[a-zA-Z0-9]*")
    return pattern.sub("",text)
```

```python
data['text']=data['text'].apply(remove_url)
```

```python
for i in data['text']:
  print(i)
```

Remove punctuation

```python
import string
punc=string.punctuation
punc
```

```python
def remove_punctuation(text):
    for char in punc:
        text=text.replace(char,"")
    return text
```

```python
data['text']=data['text'].apply(remove_punctuation)
```

Remove Emoji

```python
import emoji
def remove_emoji(text):
    clean_text=emoji.demojize(text)
    return clean_text
```

```python
data['text']=data['text'].apply(remove_emoji)
```

```python
for txt in data['text']:
    print(txt)
```

Remove whitespace

```python
def remove_whitespace(text):
    txt=" ".join(text.split())
    return txt
```

```python
data['text']=data['text'].apply(remove_whitespace)
```

```python
for i in data['text']:
    print(i)
```

Remove stopword

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text=[]
    for word in text.split():
        if word not in stopwords.words("english"):
            new_text.append(word)
    return " ".join(new_text)
```

```python
data['text_without_stopwords']=data['text'].apply(remove_stopwords)
```

```python
for text in data['text_without_stopwords']:
  print(text)
```

Tokenization

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
def tokenize(text):
    return word_tokenize(text)
```

```python
data['tokenize']=data['text'].apply(tokenize)
```

stemming and lemmetization

```python
from nltk.stem import PorterStemmer
```

```python
def stemming(text):
    obj=PorterStemmer()
    stem_word=[obj.stem(word) for word in text.split()]
    return stem_word
```

```python
data['stem']=data['text'].apply(stemming)
```

```python
for i in data['stem']:
    print(i)
```
